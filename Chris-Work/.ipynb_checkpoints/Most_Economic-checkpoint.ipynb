```python
import pandas as pd
df = pd.read_csv('original_frame.csv')
```

```python
df.columns.get_loc('PerformanceRating')
economic_df = df[['Year',
                  'Make',
                  'Model',
                  'ConsumerReviews',
                  'PerformanceRating', 
                  'ValueForMoneyRating', 
                  'ReliabilityRating', 
                  'MaxMPG', 
                  ]]
```

```python
economic_df
```

```python
model_group = economic_df.groupby(economic_df['Model']).mean()
```

```python
model_group
```

```python
highest_performer = model_group[(model_group['PerformanceRating']>= 4.75) & 
                                (model_group['ValueForMoneyRating']>=4.75)&
                                (model_group['ReliabilityRating'] >= 4.5) &
                               (model_group['MaxMPG']>= 40) &
                               (model_group['ConsumerReviews']>= 50)]
```

```python
highest_performer
# cars with Perfroamcne and Value Ratings higher than 4.75/5, Reliability higher than 4.5/5, more than 40 mpg, and at least 50 avg num reviews.
```

```python
highest_performer.plot.bar( y= 'MaxMPG', color = 'red',
                           figsize = (7,7),
                           title = 'Top Peforming Cars MPG'
                          )
```

```python
highest_performer.plot.bar( y = 'ConsumerReviews', color = 'purple',
                          figsize = (7,7),
                          title = '# of Consumer Reviews for Most Economic Cars',
                          )
```
